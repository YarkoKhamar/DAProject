```python
import numpy as np
import pandas as pd 
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('CompleteDataset.csv')
```

```python
teams = pd.read_csv('full_team_data.csv')
```

```python
teams.drop('Unnamed: 0', 1,  inplace=True)
```

```python
data.drop('Unnamed: 0', 1, inplace=True)
```

```python
def change_cols_order(cols, from_list, after_col):
    cols = np.delete(cols,[cols.index(col) for col in from_list]).tolist();
    after_col_ind = cols.index(after_col)+1;
    return np.insert(np.array(cols, dtype="O"), after_col_ind, from_list).tolist();
```

```python
data['Club'].replace({'Minnesota Thunder': 'Minnesota United'}, inplace=True)
```

```python
merged_data = pd.merge(data, teams, how = 'left', left_on = 'Club', right_on = 'Club')
merged_data.columns
```

```python
cols = merged_data.columns.tolist()
cols = change_cols_order(cols,['Preferred Positions'],'Volleys')
cols = change_cols_order(cols, teams.columns[1:],'Club')
merged_data = merged_data[cols]
merged_data.set_index("ID", inplace=True)
```
