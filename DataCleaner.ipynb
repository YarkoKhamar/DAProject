```python
import numpy as np
import pandas as pd 
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
```

```python
# read data into new pandas DataFrame.
data = pd.read_csv('complete_players_data.csv')
```

```python
data
```

```python
# initial number of rows
len(data.index)
```

```python
# check, whether there are any NaN values in the dataset.
data.isnull().values.any()
```

```python
# count how many NaN values are there in every column (omitting 0 quantities).
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

#### column *Club* has 248 NaN values, which means, that there are 248 free agents (players, currently without a club)

```python
# print players, that are free agents.
print(data[data["Club"].isnull()]["Name"])
```

```python
# drop rows (by index), that contain free agents.
rows_to_drop = data[data["Club"].isnull()]["Name"].index.tolist();
data.drop(rows_to_drop,inplace=True)
```

```python
# check, if done correctly.
data[null_columns].isnull().sum()
```

#### as we can see, all rows with free agents have been dropped.

```python
# the number of rows is reduced accordingly.
len(data)
```

```python
# find players with empty name field (if any)
counter = 0;
for row in data["Name"]:
    if row == "":
        counter += 1;
counter
```

#### so, there aren't any players with empty string as a name in our dataset.

```python
# print(data["Preferred Positions"])
# convert "Preferred Positions" column values to single position and remove whitespace at the end.
data["Preferred Positions"].replace(regex={r'^(\w+).*$':r'\1'}, inplace=True)
```

```python
data["Preferred Positions"]
```

```python
len(data["Preferred Positions"])
```

```python
# Check if 1987 NaN values for playing position columns are GK (goalkeepers), since they only play one position.
# So, count number of rows, for which the value of "Preferred Positions" column is "GK"
gk_match = data['Preferred Positions'] == "GK"
len(data[gk_match])
```

#### so, the number of GKs is exactly 1987. Since they don't play any other position, NaN values they have in other positions columns need not be filled with something else. In fact, they are handled in a proper way by pandas methods.

```python
# drop columns of the dataset, that are not useful for analysis purposes
data.drop(columns = ['Photo','Flag','Club Logo','Special'], inplace=True)
```

```python
print(data.columns)
[data[col].dtype for col in data.columns]
```

```python
# change dtype of "League Level" to integer.
data["League Level"] = data["League Level"].astype(int)
```

```python
# convert wage and value column values to numbers.
def convert_currency(curr_col):
    without_euro_symbol = curr_col.str[1:]
    unit_symbol = without_euro_symbol.str[-1]
    numeric_part = np.where(unit_symbol == '0', 0, without_euro_symbol.str[:-1].pipe(pd.to_numeric))
    multipliers = unit_symbol.replace({'M': 1e6, 'K': 1e3}).pipe(pd.to_numeric)
    return numeric_part * multipliers
```

```python
# "Wage" column is now a floating point number.
data["Wage"] = convert_currency(data["Wage"])
```

```python
# "Value" column is now a floating point number.
data["Value"] = convert_currency(data["Value"])
```

```python
# columns Acceleration ... Volleys have dtype object. We will convert them to float.
pd.DataFrame({"Column":data.columns,"DType":[data[col].dtype for col in data.columns]})
```

```python
# Example of wrong data in mentioned above columns (Dribbling = 78+3 => 78)
data[data["ID"]==189513].T
```

```python
# Correct attributes representation and convert dtype of attribute columns (Acceleration ... Volleys) to integer.
for col in data.columns[13:47]:
    data[col].replace(regex={r'^(\d+)[\+\-].*$':r'\1'}, inplace=True)
    data[col] = pd.to_numeric(data[col])
pd.DataFrame({"Column":data.columns,"DType":[data[col].dtype for col in data.columns]})
```

```python
# check if ID is unique, if so, set it as an index
data["ID"].is_unique
```

#### so, we now delete the duplicates

```python
data.drop_duplicates('ID', keep='first', inplace=True)
```

```python
# confirm, there are no duplicates
print(data[data["ID"].duplicated()])
data["ID"].is_unique
```

```python
# check column "ID" as an index
data.set_index("ID")
```

```python
# confirm data integrity
data[null_columns].isnull().sum()
```

#### so, data integrity is confirmed (number of GK's reduced, courtesy of duplicates)
