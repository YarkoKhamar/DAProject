```python
import numpy as np
import pandas as pd 
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline

import random
import urllib.request
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
base_url = "https://sofifa.com/teams/club?offset=0"
offset = 0
columns = ['Club', 'League', 'League Level', 'Country', 'Region']
data_teams = DataFrame(columns=columns)
for offset in range(9):
    url = base_url + str(offset*80)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    table_body = soup.find('tbody')
    counter = 0
    for row in table_body.findAll('tr'):
        td = row.findAll('td')
        club = td[1].find('div').findAll('a')[0].text
        league = td[1].find('div').findAll('a')[1].text.split('(')[0]
        league_level = td[1].find('div').findAll('a')[1].text.split('(')[1].split(')')[0]
        country = td[2].find('div').findAll('a')[0].text
        region = td[2].find('div').findAll('a')[1].text
        club_data = DataFrame([[club, league, league_level, country, region]])
        club_data.columns = columns
        data_teams = data_teams.append(club_data, ignore_index=True)
        counter+=1
    offset+=1
    print(offset)
    data_teams.to_csv('full_team_data.csv', encoding='utf-8')
```
